## Dealing with Categorical Features

To apply Machine Learning algorithm, we need to convert categorical variable into some indicator variables.
For example, in diagnosis column for M/B, we need to input 0/1 to make the Machine Learning model work. For this, we will create a new column indicating the diagnosis as 0/1 for M/B, this process is called creating dummies. This is important because the Machine Learning model will not take string variable as input to work with.


```python
data.diagnosis = [1 if value =='M' else 0 for value in data.diagnosis]
```
